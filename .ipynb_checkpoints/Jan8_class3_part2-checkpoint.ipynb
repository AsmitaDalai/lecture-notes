**Loops:**  Purpose of these is to teach computer to **do repetitive tasks**.  The loop construct in Python allows you to repeat a body of code several times. 

There are two types of loops - definite loops and indefinite loops. 

Both of these structures use an **iteration variable**.

**Indefinite Loops:** This is a type of loop that will continue to run infinite number of times until and unless it is asked to stop. 

In order to execute an indefinite loop, we use the WHILE statement. The statement includes a question that gets evaluated **before** each iteration and it will run the underlying indented block as long as the answer to the question remain False.

```python
n = 5
while n > 0 :
    print(n)
    n = n - 1
print('Done!')
print(n)
```

```python
n = 5
while n > 0 :
    print('DKU')
print('Done!')
```

```python
while True:
    line = input('> ')
    if line == 'done' :
        break
    print(line)
print('Done!')
```

**Definite Loops:** You use a definite loop when you know a priori how many times you will be executing the body of the loop. These structures use the FOR and IN key words.

```python
for i in [5, 4, 3, 2, 1] :
    print(i)
print('finished!')
```

**Range(n) function**

Return an object that produces a sequence of integers from start (inclusive)to stop (exclusive) by step.  range(i, j) produces i, i+1, i+2, ..., j-1. start defaults to 0, and stop is omitted!  

range(4) produces 0, 1, 2, 3.
These are exactly the valid indices for a list of 4 elements.
When step is given, it specifies the increment (or decrement).



```python
for i in  range (5):
    print(i)
```

**Idioms:** These are **segments of code** representing a particular logical pattern of doing things.  They are handy and frequently used. It follows the following basic structure:

* Initialize iteration variable before loop
* As long as things in set, Do something 
* Then get payoff

Other idiom examples: Finding smallest number, Finding largest number – use largest so far variable – , Counting, Summing values, Filtering in loop, Using Boolean variable, None type is stronger than equal as it requires equal in value and type (ask for example)

```python
# Finding largest value on a list
largest_so_far = -1 # initializes variable       
print('Before', largest_so_far)
for the_num in [9, 41, 12, 3, 74, 15] :
    if the_num > largest_so_far :
        largest_so_far = the_num
    print('largest so far',largest_so_far, 'current num',the_num)
print('After', largest_so_far)
```

```python
# Calculating sum of items on a list
mysum = 0
print('Before', mysum)
for number in [9, 41, 12, 3, 74, 15] :
    mysum = mysum + number
    print(mysum, number)
print('After', mysum)
```

```python
# Filtering in a Loop
print('Before') 
for value in [9, 41, 12, 3, 74, 15] :
    if value > 20:
        print('Large number',value)
print('After')
```

```python
# Finding a particular value on list (e.g. 3)
found = False
print('Before', found)
for value in [9, 41, 12, 3, 74, 15] : 
    if value == 3 :
        found = True
    print(found, value)
print('After', found)
```

```python
# Here is a much easier way to implement idiom above
3 in [9, 41, 12, 3, 74, 15]
```

### Creating a list of number using a definite loop

Say you are trying to create a sequence (list) of n numbers starting with one, where each element is the square of its index

```python
n=5
mylist = [i**2 for i in range (1,n+1)]
```
